# <a name="HW4Standard">The Standard Basis for $\mathbb{P}_n$</a>

---

Recall, the <font color="dodgerblue">**standard basis**</font> for $\mathbb{P}_n$ is the set $\{1,x,x^2,\dots,x^n\}$ since it is most natural for us to express a degree $n$ polynomial as a linear combination of elements in $\{1,x,x^2,\dots,x^n\}$,

$$p(x) = a_0 + a_1 x + a_2 x^2 + \ldots + a_n x^n.$$




# <a name="HW4Inter-general">Fitting a Polynomial Model to $n+1$ Data Points<a>

---

Thus far, we have noted:

- Given 2 points, there is a unique linear function (degree 1 polynomial) whose graph passes through both points.
- Given 3 data points, there is a unique quadratic function (degree 2 polynomial) whose graph passes through all three points.

More generally:

- Given $n+1$ points $(x_0,y_0),(x_1,y_1),\dots,(x_n,y_n)$ in $\mathbb{R}^2$ (with distinct $x$-values) there is a unique degree $n+1$ polynomial $p(x)$ in the vector space $\mathbb{P}_n$ that passes through all $n+1$ points.


We can apply the same interpolation method as we applied to solve [Question 1](#HW4Q1):



## Step 1:

---

Plug each point into a general formula $p(x) = a_0 + a_1 x + a_2 x^2 + \ldots + a_n x^n$ to obtain $n+1$ equations for the $n+1$ undetermined coefficients $a_0$, $a_1$, $\ldots$, $a_n$.

$$\begin{align}
p(x_0) = a_0 + a_1 x_0 + a_2 x_0^2 + \ldots + a_n x_0^2 &= y_0 \\
p(x_1) = a_0 + a_1 x_1 + a_2 x_1^2 + \ldots + a_n x_1^2 &= y_1 \\
p(x_2) = a_0 + a_1 x_2 + a_2 x_2^2 + \ldots + a_n x_2^2 &= y_2 \\
\vdots & \\
p(x_n) = a_0 + a_1 x_n + a_2 x_n^2 + \ldots + a_n x_n^2 &= y_n \\
\end{align}$$

<br>  



## Step 2:

---

Set up and solve the matrix equation $$V \mathbf{a} = \mathbf{y}$$ corresponding to the system of $n+1$ linear equations from Step 1 to find values for the coefficients $a_0$, $a_1$, $\ldots$, $a_n$.

- The coefficient matrix $V$ is the $(n + 1) \times (n + 1)$ matrix

$$V=\begin{bmatrix}
1 & x_0 & x_0^2 & \dots & x_0^n \\
1 & x_1 & x_1^2 & \dots & x_1^n \\
1 & x_2 & x_2^2 & \dots & x_2^n \\
\vdots \\
1 & x_n & x_n^2 & \dots & x_n^n \\
\end{bmatrix}.$$

- The column vector of undetermined coefficients is

$$\mathbf{a}=\begin{bmatrix} a_0 \\ a_1 \\ \vdots \\ a_n\end{bmatrix}.$$

- The column vector of constants on the right side of the equations is

$$\mathbf{y}=\begin{bmatrix} y_0 \\ y_1 \\ \vdots \\ y_n\end{bmatrix}.$$


- Note: The $(n+1)\times(n+1)$ matrix $V$ is called a <font color="dodgerblue">**Vandermonde matrix**</font>.

# <a name="HW4Newton">Another Basis: Newton Polynomials</a>

---

The **standard basis**, $\{1,x,x^2,\dots,x^n\}$, for $\mathbb{P}_n$ is perhaps the most natural way to express a polynomial in $\mathbb{P}_n$ as a linear combination of basis elements. There are other bases we can use for $\mathbb{P}_n$.

Given a data set of $n+1$ points (each with distinct $x$-values),

$$(x_0,y_0),(x_1,y_1),\dots,(x_n,y_n),$$

the <font color="dodgerblue">**Newton polynomials**</font> are defined based on $x$-coordinates of the given data points:

$$\begin{align}
N_0(x) &= 1 \\
N_1(x) &= (x-x_0) \\
N_2(x) &= (x-x_0)(x-x_1) \\
\vdots \\
N_j(x) &= (x-x_0)(x-x_1)\cdots (x-x_{j-1}) = \prod_{k=0}^{j-1}(x-x_k) \\
\vdots \\
N_n(x) &= (x-x_0)(x-x_1)\cdots (x-x_{n-1}).
\end{align}$$

The set of Newton polynomials for another basis for $\mathbb{P}_n$.  Note that the Newton polynomial:

-  $N_0(x)$ has degree $0$.
-  $N_1(x)$ has degree $1$.
- $\vdots$
-  $N_j(x)$ has degree $j$.
- $\vdots$
- $N_n(x)$ has degree $n$.





# <a name="HW4Inter-newton">Interpolation with Newton Polynomials</a>

---


We can apply a similar interpolation method [as we used with the standard basis](#HW4Inter-general) for $\mathbb{P}^n$ using the basis of Newton polynomials instead.



## Step 1:

---

Plug each point into a general formula $p(x) = b_0 N_0(x) + b_1 N_1(x) + b_2 N_2(x) + \ldots + b_n N_n(x)$ to obtain $n+1$ equations for the $n+1$ undetermined coefficients $b_0$, $b_1$, $\ldots$, $b_n$.

$$\begin{align}
p(x_0) = b_0 N_0(x_0) + b_1 N_1(x_0) + \ldots + b_n N_n(x_0) &= y_0 \\
p(x_1) = b_0 N_0(x_1) + b_1 N_1(x_1) + \ldots + b_n N_n(x_1) &= y_1 \\
p(x_2) = b_0 N_0(x_2) + b_1 N_1(x_2) + \ldots + b_n N_n(x_2) &= y_2 \\
\vdots & \\
p(x_n) = b_0 N_0(x_n) + b_1 N_1(x_n) + \ldots + b_n N_n(x_n) &= y_n \\
\end{align}$$

<br>  



## Step 2:

---

Set up and solve the matrix equation $$M \mathbf{b} = \mathbf{y}$$ corresponding to the system of $n+1$ linear equations from Step 1 to find values for the coefficients $b_0$, $b_1$, $\ldots$, $b_n$.

- The coefficient matrix $M$ is the $(n + 1) \times (n + 1)$ matrix

$$M=\begin{bmatrix}
N_0(x_0) & N_1(x_0) & N_2(x_0) & \dots & N_n(x_0) \\
N_0(x_1) & N_1(x_1) & N_2(x_1) & \dots & N_n(x_1) \\
N_0(x_2) & N_1(x_2) & N_2(x_2) & \dots & N_n(x_2) \\
\vdots \\
N_0(x_n) & N_1(x_n) & N_2(x_n) & \dots & N_n(x_n) \\
\end{bmatrix}.$$

- The column vector of undetermined coefficients is

$$\mathbf{b}=\begin{bmatrix} b_0 \\ b_1 \\ \vdots \\ b_n\end{bmatrix}.$$

- The column vector of constants on the right side of the equations is still

$$\mathbf{y}=\begin{bmatrix} y_0 \\ y_1 \\ \vdots \\ y_n\end{bmatrix}.$$

# <a name="HW4Newton">Yet Another Basis: Lagrange Polynomials</a>

---

Like the Newton polynomials, the <font color="dodgerblue">**Lagrange polynomials**</font> depend on the $x$-coordinates of the $n+1$ given data points

$$(x_0,y_0),(x_1,y_1),\dots,(x_n,y_n).$$


The Lagrange polynomials defined below form another basis for $\mathbb{P}^n$.

$$\begin{align}
L_0(x) &= (x-x_1)(x-x_2)\cdots(x-x_n) \\
L_1(x) &= (x-x_0)(x-x_2)\cdots(x-x_n) \\
L_2(x) &= (x-x_0)(x-x_1)(x-x_3)\cdots(x-x_n) \\
\vdots \\
L_j(x) &= (x-x_0)\cdots (x-x_{j-1})(x-x_{j+1})\cdots(x-x_n) = \prod_{\substack{k=0}{k\ne j}}^{n}(x-x_k) \\
\vdots \\
L_n(x) &= (x-x_0)(x-x_1)\cdots (x-x_{n-1})
\end{align}$$


- <font color="dodgerblue">Note that all Lagrange polynomials in this basis have degree $n$.</font>



# <a name="HW4Inter-lagrange">Interpolation with Lagrange Polynomials</a>

---


We can apply a similar interpolation method [as we used with the standard basis](#HW4Inter-general) for $\mathbb{P}^n$ using the basis of Lagrange polynomials instead.


## Step 1:

---

Plug each point into a general formula $p(x) = c_0 L_0(x) + c_1 L_1(x) + c_2 L_2(x) + \ldots + c_n L_n(x)$ to obtain $n+1$ equations for the $n+1$ undetermined coefficients $c_0$, $c_1$, $\ldots$, $c_n$.

$$\begin{align}
p(x_0) = c_0 L_0(x_0) + c_1 L_1(x_0) + \ldots + c_n L_n(x_0) &= y_0 \\
p(x_1) = c_0 L_0(x_1) + c_1 L_1(x_1) + \ldots + c_n L_n(x_1) &= y_1 \\
p(x_2) = c_0 L_0(x_2) + c_1 L_1(x_2) + \ldots + c_n L_n(x_2) &= y_2 \\
\vdots & \\
p(x_n) = c_0 L_0(x_n) + c_1 L_1(x_n) + \ldots + c_n L_n(x_n) &= y_n \\
\end{align}$$

Set up and solve the matrix equation $$K \mathbf{b} = \mathbf{y}$$ corresponding to the system of $n+1$ linear equations from Step 1 to find values for the coefficients $c_0$, $c_1$, $\ldots$, $c_n$.

$$K=\begin{bmatrix}
L_0(x_0) & L_1(x_0) & L_2(x_0) & \dots & L_n(x_0) \\
L_0(x_1) & L_1(x_1) & L_2(x_1) & \dots & L_n(x_1) \\

L_0(x_n) & L_1(x_n) & L_2(x_n) & \dots & L_n(x_n) \\
\end{bmatrix}.$$

- The column vector of undetermined coefficients is

$$\mathbf{c}=\begin{bmatrix} c_0 \\ c_1 \\ \vdots \\ c_n\end{bmatrix}.$$

- The column vector of constants on the right side of the equations is still

$$\mathbf{y}=\begin{bmatrix} y_0 \\ y_1 \\ \vdots \\ y_n\end{bmatrix}.$$

## <a name="HW4guided">A Guided Example</a>

---

Suppose we would like to solve the following system of equations:

$$\begin{align}
2 a_0 + 4 a_1 + 6 a_2 + 8 a_3  &= 6 \\
5 a_0 + 7 a_1 + 9 a_2 + 11 a_3 &= -2 \\
6 a_0 + 8 a_1 + 10 a_2 + 12 a_3 &= -5 \\
11 a_0 + 13 a_1 + 15 a_2 + 17 a_3 &= 10
\end{align}$$

This gives rise to the matrix equation

$$\begin{bmatrix}
2 & 4 & 6 & 8 \\
5 & 7 & 9 & 11 \\
6 & 8 & 10 & 12\\
11 & 13 & 15 & 17
\end{bmatrix}
\begin{bmatrix} a_0 \\ a_1 \\ a_2 \\ a_3 \end{bmatrix} =
\begin{bmatrix} 6 \\ -2 \\ -5 \\ 10 \end{bmatrix}$$

Do you notice a pattern in the entries of the coefficient matrix?

- Add $2$ the entries in column 1 to get the entries in column 2.
- Add $4$ the entries in column 1 to get the entries in column 3.
- Add $6$ the entries in column 1 to get the entries in column 4.





### <a name="HW4col1">The Values in Column 1</a>

---

We first define a column vector we name $\mathbf{x}$ which is equal to the first column of the coefficient matrix:

$$\mathbf{x} = \begin{bmatrix} 2 \\ 5 \\ 6 \\ 11 \end{bmatrix}.$$

- We first store the column vector $\mathbf{x}$ as a $4 \times 1$ matrix in the code cell below.
- We will eventually refer to `x` when creating a general formula for entries in $N$.

```python
import sympy as sym

x = sym.Matrix([[2],
            [5],
            [6],
            [11]])

x  # print to screen
```

### <a name="HW4index">Python Indexing</a>

Recall Python indexing starts at 0. Thus:

- `x[1, 0]` gives the entry in row 2 column 1, which is 5.
- `x[2, 0]` gives the entry in row 3 and column 1 which is 6.
- `x[3, 0]` gives the entry in row 4 and column 1 which is 11.


```python
x[1, 0]  # verify the result is 5
```

## <a name="HW4lambda">Lambda Functions in Python</a>

---

A [**lambda function**](https://www.w3schools.com/python/python_lambda.asp) can be used to generate the output of a an expression which depends on one or more inputs (or arguments). To enter a lambda function in Python:

- Begin the function by typing `lambda`.
- Then list the inputs as a list separated by commas.
- After the comma, enter formula for determining the output.


For example, we can define the function

$$f(i, j) = i + 5j^2$$

using lambda functions in the code cell below.

We can name lambda functions and evaluate the function at given inputs. For example, `f(1, 2)` computes the output

$$f(1, 2) = 1 + 5(2)^2 = 21.$$


```python
f = lambda i, j: i + 5 * j**2  # notice exponents use ** and not ^
f(1, 2)  # compute output at (1, 2)
```

### <a name="HW4guide-lambda">Creating a Lambda Function for Matrix Entries</a>

---

The system of system equations in our example has a $4 \times 4$ coefficient matrix we denote $N$.

$$N = \begin{bmatrix}
2 & 4 & 6 & 8 \\
5 & 7 & 9 & 11 \\
6 & 8 & 10 & 12\\
11 & 13 & 15 & 17
\end{bmatrix} =
\begin{bmatrix}
2 & 2+1(2) & 2 + 2(2) & 2 + 3(2)\\
5 & 5+1(2) & 5 + 2(2) & 5 + 3(2)\\
6 & 6+1(2) & 6 + 2(2) & 6 + 3(2)\\
11 & 11+1(2) & 11 + 2(2) & 11 + 3(2)
\end{bmatrix}$$

The first column of the matrix has already been stored in the column vector `x`. Thus, we have


$$N = \begin{bmatrix}
2 & 4 & 6 & 8 \\
5 & 7 & 9 & 11 \\
6 & 8 & 10 & 12\\
11 & 13 & 15 & 17
\end{bmatrix} =
\begin{bmatrix}
x[0,0] & x[0,0] + 1(2) & x[0,0] + 2(2) & x[0,0] + 3(2)\\
x[1,0] & x[1,0] + 1(2) & x[1,0] + 2(2) & x[1,0] + 3(2)\\
x[2,0] & x[2,0] + 1(2) & x[2,0] + 2(2) & x[2,0] + 3(2)\\
x[3,0] & x[3,0]+1(2) & x[3,0] + 2(2) & x[3,0] + 3(2)
\end{bmatrix}$$

**Python is indexed at 0**, so entries in:

- Column 2 of $N$ have column index `j=1`.
- Column 3 of $N$ have column index `j=2`.
- Column 4 of $N$ have column index `j=3`.
- <font color="dodgerblue">The column indices are exactly the multiples in front of 2!</font>
- Similarly, we can use the row indices `i` to identify the first entries in each row of $N$.
- Thus, the entry in row `i` and column `j` of matrix $N$ is equal to `N(i, j) = x[i, j] + j * 2`

In the code cell we check our formula. The entry in row 2 and column 3 of $N$ is $9$. Since Python indexing begins at 0, this entry has row index `i=1` and column index `j=2`.


```python
N = lambda i, j: x[i, 0] + j * 2
N(1, 2)  # check entry in row 2 and column 3
```

### <a name="HW4lambda-matrix">Creating a Matrix with Lambda Functions</a>

---

Finally, we are ready to define the coefficient matrix $N$ using a the lambda function we just identified using the following code

`sym.Matrix(4, 4, lambda i, j: x[i, 0] + j * 2)`

- The first `4` indicates $N$ has 4 rows.
- The second `4` indicates $N$ has 4 columns.
- Then we enter the lambda function that gives each entry.
  - We have already imported SymPy as `sym`.
  - We have already stored the column vector `x`.
- Run the code cell below to check the output.

```python
N = sym.Matrix(4, 4, lambda i, j: x[i, 0] + j * 2)  # enter a formula
N  # print N to check
```
