# <a name="HW6about">About the Data</a>

---



Attribute Information:

- `age`, age in years.
- `sex`, some observations are coded as `1` and others `2` (not clear which is which sex).
- `bmi`, [body mass index](https://www.cdc.gov/healthyweight/assessing/bmi/index.html).
- `bp`, [diastolic blood pressure](https://www.heart.org/en/health-topics/high-blood-pressure/understanding-blood-pressure-readings) in millimeters of mercury (mm Hg).
- `tc`, [total serum cholesterol](https://mayoclinic.org/tests-procedures/cholesterol-test/about/pac-20384601) is the sum of LDL, HDL and 20% of the triglycerides level in mg/dL.
- `ldl`, [low-density lipoproteins](https://mayoclinic.org/tests-procedures/cholesterol-test/about/pac-20384601) in mg/dL.
- `hdl`, [high-density lipoproteins](https://mayoclinic.org/tests-procedures/cholesterol-test/about/pac-20384601)  in mg/dL.
- `tch`, is the total serum cholesterol divided by HDL.
- `ltg`, natural log of serum [triglycerides level](https://my.clevelandclinic.org/health/articles/11117-triglycerides) in mg/dL.
  - Note we take the natural log of the triglycerides in order to transform the values to be more normally distributed.
- `glu`, [blood sugar level](https://my.clevelandclinic.org/health/diagnostics/12363-blood-glucose-test) in mg/dL.
- `target`, a quantitative measure of disease progression one year after baseline.
  - The greater the value of `target`, the more severe the progression of diabetes.



For more information see:
Bradley Efron, Trevor Hastie, Iain Johnstone and Robert Tibshirani (2004) "Least Angle Regression," *Annals of Statistics*, 407-499. [Link to full article](https://web.stanford.edu/~hastie/Papers/LARS/LeastAngle_2002.pdf)
